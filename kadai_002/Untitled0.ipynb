```python
# 必要なライブラリをインポート
import pandas as pd

# Step 1: 売上データを辞書形式で定義
data = {
    "日付": ["2023-05-17", "2023-05-18", "2023-05-19", "2023-05-20", "2023-05-21"],
    "社員名": ["山田", "佐藤", "鈴木", "田中", "高橋"],
    "売上": [100, 200, 150, 300, 250],
    "部門": ["メーカー", "代理店", "メーカー", "商社", "代理店"]
}

# Step 2: pandasでデータフレームを作成
df = pd.DataFrame(data)

# Step 3: 平均売上を算出し「平均売上」列を追加
average_sales = df["売上"].mean()
df["平均売上"] = average_sales

# Step 4: 業績ランクを判定する関数を定義
def performance(sales, avg):
    if sales >= avg + 50:
        return "A"
    elif sales >= avg:
        return "B"
    else:
        return "C"

# Step 5: 業績ランク列を追加
df["業績ランク"] = df["売上"].apply(lambda x: performance(x, average_sales))

# Step 6: Excelファイル「業績.xlsx」に書き出し
df.to_excel("業績.xlsx", index=False)

# 完了メッセージ
print("Excelファイル『業績.xlsx』が正常に作成されました。")
```
